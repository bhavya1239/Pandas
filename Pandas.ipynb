```python
import pandas as pd
```

```python
a = [3,4,5,6,7,8]
var = pd.Series(a)
print(var)
print(type(var))
```

```python
#index number
a = [3,4,5,6,7,8]
var = pd.Series(a)
print(var)
print(type(var))
print(var[2])
```

```python
a = [3,4,5,6]
var = pd.Series(a,index=['a','s','d','f'],dtype="float")
print(var)
print(type(var))
print(var[2])
```

```python
#name parameter
a = [3,4,5,6]
var = pd.Series(a,index=['a','s','d','f'],dtype="float",name="python")
print(var)
print(type(var))
print(var[2])
```

```python
#passing dic
dic = {"name":['python','c','java'],"por":[12,13,14,15],"rank":[1,4,3,2]}
var1 = pd.Series(dic)
#var1               this can also be written
#print()
print(var1)
```

```python
s= pd.Series(12)

print(s)
print(type(s))
```

```python
s= pd.Series(12,index=[1,2,3,4,5,6,7])

print(s)
print(type(s))
```

```python
#missing data
s= pd.Series(12,index=[1,2,3,4,5,6,7])
s1= pd.Series(12,index=[1,2,3,4])

print(s + s1 )
```

```python
#ARITHMETIC OPERATIONS
```

```python
qa = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
qa
```

```python
#ADDITION
qa["C"] = qa["A"] +qa["B"]
qa
```

```python
#SUBTRACTION
qa["C"] = qa["A"] -qa["B"]
qa
```

```python
#MULTIPLY
qa["C"] = qa["A"] *qa["B"]
qa
```

```python
#Divide
qa["C"] = qa["A"] /qa["B"]
qa
```

```python
#Divide
qa["C"] = qa["A"] //qa["B"]
qa
```

```python
qa1 = pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
qa1
```

```python
qa1["Python"] = qa1["A"] <= 20
qa1
```

```python
qa1["Python"] = qa1["A"] <= 20
qa1["Python_1"] = qa1["B"] >= 16
qa1
```
